```python
import pandas as pd
import matplotlib.pyplot as plt
```

**LINE CHART - growth/decline**

```python
data={
    "Month":["Jan","Feb","Mar","Apr","May","Jun"],
    "Sales":[20000,25000,18000,30000,29000,21000]
}
```

```python
df=pd.DataFrame(data)
print(df)
```

```python
plt.plot(df["Month"],df["Sales"],marker='o')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Sales in 2025")
plt.show()
```

**BAR CHART - compare values**

```python
products=["Laptop","Mobile","Tablet","SmartWatch"]
sales=[50,80,30,45]

plt.bar(products,sales)
plt.xlabel("Product")
plt.ylabel("Units sold")
plt.title("Sales for Nov 2025")
plt.show()
```

**PIE CHART - percentage contribution**

```python
sales=[50,80,30,45]
products=["Laptop","Mobile","Tablet","SmartWatch"]

plt.pie(sales,labels=products,autopct='%1.1f%%')
plt.title("Sales Share Productwise")
plt.show()
```

**SCATTER CHART- identifies correlation**

```python
study_hours=[1,2,3,4,5,6]
marks=[40,50,60,70,80,90]

plt.scatter(study_hours,marks)
plt.xlabel("study hours")
plt.ylabel("marks")
plt.title("study hours vs marks")
plt.show()
```

BOX CHART - outliers

```python
import pandas as pd

data = {
    "Month": [
        "Jan", "Feb", "Mar", "Apr", "May", "Jun",
        "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
    ],
    "Sales": [
        9500, 10200, 9800, 10500, 9900, 10100,
        9700, 10300, 50000, 10000, 9900, 10200
    ]
}

df = pd.DataFrame(data)
print(df)
```

```python
plt.boxplot(df['Sales'])
plt.title("Sales distribution")
plt.ylabel("Sales")
plt.show()
```
