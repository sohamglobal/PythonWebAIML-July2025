```python
import pandas
```

**Load the date**

```python
url="https://raw.githubusercontent.com/sohamglobal/datasets/refs/heads/main/Flipkart_Mobiles.csv"
df=pandas.read_csv(url)
print(df)
```

**Top 5 performing brands**

```python
top_brands=df['Brand'].value_counts().head(5)
print(top_brands)
```

**Top 10 Brand percentage share**

```python
brand_count=df['Brand'].value_counts()
print(brand_count.sum())
```

```python
print(brand_count)
```

```python
brand_perc=(brand_count/brand_count.sum())*100
brand_distribution=pandas.DataFrame({'Count':brand_count,'Percentage':brand_perc})
print(brand_distribution.head(10))
```

**Mobile sale on Color**

```python
top_colors=df['Color'].value_counts().head(10)
print(top_colors)
```

**Top rates 10 Mobiles**

```python
top_rated_mobiles=df[['Model','Brand','Rating']].sort_values(by='Rating',ascending=False)
print(top_rated_mobiles.head(10))
```

**Mobiles with highest discount**

```python
df['Discount']=df['Original Price']-df['Selling Price']
print(df)
```

```python
top_discount=df[['Model','Brand','Original Price','Discount','Selling Price']].sort_values('Discount',ascending=False)
print(top_discount.head(10))
```

**Lowest rated mobiles**

```python
lowest_rated=df[['Model','Brand','Rating']].sort_values('Rating').head(10)
print(lowest_rated)
```
