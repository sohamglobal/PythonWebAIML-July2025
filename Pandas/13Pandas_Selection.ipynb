**Retrieve data from the pandas dataframe on the basic of row and columns**

```python
import pandas
```

```python
url="http://sohamspring.azurewebsites.net/api/accounts/all"
df=pandas.read_json(url)
print(df)
```

```python
#retrieve data from a specific column
print(df['accnm'])
```

```python
#retrieve data from multiple columns
print(df[['accnm','balance']])
```

```python
#collect data from a specific row
print(df.loc[3])
```

```python
#retrieve data of multiple rows
print(df.loc[5:9])
```

```python
#retrieve column value of spec row
print(df.loc[13,'accnm'])
```

```python
#retrieve multiple column values of a spec row
print(df.loc[15,['accnm','acctype','balance']])
```

```python
#retrieve data of multiple rows and single column
print(df.loc[11:17,'accnm'])
```

```python
#retrieve multiple rows of multiple columns
print(df.loc[3:9,['accnm','balance']])
```
