```python
import pandas
```

```python
url1="https://raw.githubusercontent.com/sohamglobal/datasets/refs/heads/main/doctors.csv"
dfdoc=pandas.read_csv(url1)
print(dfdoc)
```

```python
url2="https://raw.githubusercontent.com/sohamglobal/datasets/refs/heads/main/patients.csv"
dfpat=pandas.read_csv(url2)
print(dfpat)
```

```python
res=dfpat.merge(dfdoc,on="doctor_id",how="inner")
print(res[['patient_name','ailment','doctor_name','specialization']])
```

```python
res=dfpat.merge(dfdoc,on="doctor_id",how="left")
print(res[['patient_name','ailment','doctor_name','specialization']])
```

```python
res=dfpat.merge(dfdoc,on="doctor_id",how="right")
print(res[['patient_name','ailment','doctor_name','specialization']])
```
