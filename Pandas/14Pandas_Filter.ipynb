**Retrieve data on the basis of certain conditions**

```python
import pandas
```

```python
url="http://sohamspring.azurewebsites.net/api/accounts/all"
df=pandas.read_json(url)
print(df)
```

```python
print(df[df['acctype']=='current'])
```

```python
print(df[df['acctype']!='saving'])
```

```python
print(df[(df['acctype']=='fixed') & (df['balance']<50000)])
```

```python
print(df[(df['accno']==1002) | (df['balance']>95000)])
```

```python
print(df[df['accno'].isin([1027,1014,1022,1035,1029])])
```

```python
res=df[df['balance']<25000]
print(res)
```
