```python
url="http://sohamspring.azurewebsites.net/api/orders/all"
df=pandas.read_json(url)
print(df)
```

```python
res=df.sort_values('company')
print(res)
```

```python
res=df.sort_values('city')
print(res)
```

```python
res=df.sort_values('price',ascending=False)
print(res)
```
