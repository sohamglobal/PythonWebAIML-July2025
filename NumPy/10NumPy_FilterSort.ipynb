```python
import numpy
```

**Fliter**

```python
a1=numpy.random.randint(1,75,10)
print(a1)
```

```python
filtered=a1[a1>50]
print(filtered)
```

```python
odd=a1[a1%2==1]
print(odd)
```

**Sort**

```python
a2=numpy.random.randint(25,100,10)
print(a2)
```

```python
print(numpy.sort(a2))
```

```python
print(numpy.sort(a2)[::-1])
```

**Unique**

```python
a3=numpy.array([9,26,1,13,30,45,9,25,1,17,26,15,13])
print(numpy.unique(a3))
```

**Split**

```python
a4=numpy.array([9,26,1,13,30,45,9,25,1,17,26,15,13,11,34])
print(numpy.split(a4,5))
```

```python
a5=numpy.array([9,26,1,13,30,45,9,25,1,17,26,15])
print(a5.shape)
print(numpy.array_split(a5,5))
```
