NumPy is the base library for all data science operations. It is fast and performs data processing effectively.

```python
import numpy
```

**Creating numpy array from a 1-D list**

```python
lst=[9,26,13,1,30,12,44,89,31]
print(lst)
print(type(lst))
```

```python
arr1=numpy.array(lst)
print(arr1)
print(type(arr1))
```

**Create numpy array from 2-D list**

```python
lst2=[
    [12,45,32,56],
    [25,78,81,42],
    [11,78,43,22]
]
print(lst2)
```

```python
arr2=numpy.array(lst2)
print(arr2)
```

**processing a list**

```python
for row in lst2:
  for val in row:
    val+=3
    print(val)
```

**processing numpy array**

```python
arr2+=3
print(arr2)
```

```python
arr2=arr2**2
print(arr2)
```

**Creating numpy array using 0's and 1's**

```python
z=numpy.zeros((4,5))
print(z)
```

```python
o=numpy.ones(9)
print(o)
```

```python
narr=numpy.full((3,5),13)
print(narr)
```

**creating identity matrix**

```python
iarr=numpy.eye(5)
print(iarr)
```

**create numpy array using arange()**

```python
a1=numpy.arange(1,25)
print(a1)
```

```python
a2=numpy.arange(75,90,2)
print(a2)
```

```python
a3=numpy.arange(50,25,-3)
print(a3)
```

**creating numpy array using linear spacing**

```python
a4=numpy.linspace(0,1,5)
print(a4)
```

```python
a5=numpy.linspace(1,13,9)
print(a5)
```
