**create numpy array with random values**

```python
import numpy
```

```python
arr1=numpy.random.random(10)
print(arr1)
```

```python
arr2=numpy.random.random((3,5))
print(arr2)
```

```python
arr3=numpy.random.randint(10)
print(arr3)
```

```python
arr4=numpy.random.randint(1,50,10)
print(arr4)
```

```python
arr5=numpy.random.randint(1,100,(5,7))
print(arr5)
```

```python
print(arr5.shape)
```

**Create a numpy array with lambda function**

```python
f=lambda n:n**2
```

```python
print(f(13))
```

```python
f=lambda i,j:i+j
print(f(9,13))
```

```python
arr6=numpy.fromfunction(lambda i,j:i+j,(3,3),dtype=int)
print(arr6)
```
