```python
import random
```

```python
# Actions the dog can take
actions=["sit","stand"]
```

```python
# Allot memory for rewards for each action
reward_table={
    "sit":0,
    "stand":0
}
```

```python
# Training loop
for episode in range(1,21):
  # Dog chooses action randomly
  action=random.choice(actions)

  # Give reward
  if action=="sit":
    reward=1  # biscuit
  else:
    reward=0  # no reward

  # Dog updates the learning memory
  reward_table[action]+=reward

  print(f"Episode {episode}: Action : {action}, Reward: {reward}")

print("\nLearning result (Reward table) :")
print(reward_table)
```

```python
best_action=max(reward_table, key=reward_table.get)
print("\nDog has lerned to : ", best_action)
```
