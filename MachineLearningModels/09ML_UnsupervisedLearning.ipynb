```python
import pandas
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
url="https://raw.githubusercontent.com/sohamglobal/datasets/refs/heads/main/testcricketers.csv"
df=pandas.read_csv(url)
print(df)
```

```python
X=df[['TestRuns','TestWickets']]
```

```python
km=KMeans(n_clusters=3,random_state=42)
df['Cluster']=km.fit_predict(X)
```

```python
print(df)
```

```python
print(df.groupby('Cluster')[['TestRuns','TestWickets']].mean())
```

```python
plt.scatter(df['TestRuns'],df['TestWickets'],c=df['Cluster'])
plt.xlabel('Runs')
plt.ylabel('Wickets')
plt.title('Unsupervised clustering of test cricketers')
plt.show()
```

```python
newplayer={'TestRuns':[4500],'TestWickets':[180]}
np=pandas.DataFrame(newplayer)
result_cluster=km.predict(np)
print(result_cluster)

if result_cluster[0]==0:
  print('Batter')
elif result_cluster[0]==1:
  print('Bowler')
else:
  print('All-rounder')
```
