```python
import pandas
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
url="https://raw.githubusercontent.com/sohamglobal/datasets/refs/heads/main/moviechoice.csv"
df=pandas.read_csv(url)
print(df)
```

```python
features=df[['gender','age','locality']]
labels=df['genre']
```

```python
total=0
for i in range(1,11):
  F_train,F_test,L_train,L_test=train_test_split(features,labels,test_size=0.2)
  model=DecisionTreeClassifier()
  model.fit(F_train.values,L_train.values)
  predicted_result=model.predict(F_test.values)
  score=accuracy_score(predicted_result,L_test)
  #print(score)
  total+=score

avg=(total/10)*100
print('Average accuracy score of the ML model : %.2f%%' %avg)

if avg>80:
  print('This model can be deployed in an application')
else:
  print('This model needs more data from training')
```
